## Clean a Scarab export so that we can render the pre-TEP more easily

The export contains redundant rows (repeats) and columns (that we wouldn't use in a TEP)

```python
import pandas as pd
```

```python
raw_export = pd.read_csv("scarab_export/scarab_query_2023_04_13.csv")
raw_export
```

```python
# keep only entries where sequence data is defined.
only_sequence_data_export = raw_export[raw_export['Construct ID'].notna()]
only_sequence_data_export
```

```python
# remove duplicates. These are usually just assay replicates, so the assay information is conserved across them.
# drop dupes by checking both virus family and target name.
uniq_export = only_sequence_data_export.drop_duplicates(subset=["Target Code", "Target Description"], keep="first")
uniq_export
```

```python
# drop columns that are all NaN.
uniq_export = uniq_export.dropna(axis=1, how='all')
```

```python
# save to csv.
uniq_export.to_csv("scarab_cleaned/scarab_export_cleaned.csv", index=False)
```
